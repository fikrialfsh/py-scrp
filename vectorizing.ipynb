```python
#library

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import pickle
```

```python
df = pd.read_csv('true_dataset.csv')
df.head()
```

```python
vectorizer = TfidfVectorizer()
response = vectorizer.fit_transform(df['tweet'])
print(response)
```

```python
vectorizer.get_feature_names()
```

```python
response.todense()
```

```python
df = pd.DataFrame(response.todense().T,
                  index=vectorizer.get_feature_names(),
                  columns=[f'D{i+1}' for i in range(len(df))])

df
```

```python
filename = 'fake_vectorized.pkl'
pickle.dump(response, open(filename, 'wb'))
```
