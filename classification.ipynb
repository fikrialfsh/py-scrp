```python
truenews = pd.read_csv('true_dataset.csv')
fakenews = pd.read_csv('fake_dataset.csv')
```

```python
truenews['Fakta/Hoaks']='Fakta'
fakenews['Fakta/Hoaks']='Hoaks'
```

```python
# Combine the 2 DataFrames into a single data frame
news = pd.concat([truenews, fakenews])
news["Artikel"] = news["name"] + news["tweet"]
news.sample(frac = 1)
```

```python
#Data Cleaning
from nltk.corpus import stopwords
import string
```

```python
def process_text(s):

    # Check string to see if they are a punctuation
    nopunc = [char for char in s if char not in string.punctuation]

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
    
    # Convert string to lowercase and remove stopwords
    clean_string = [word for word in nopunc.split() if word.lower() not in stopwords.words('indonesian')]
    return clean_string
```

```python
# Tokenize the text :Convert the normal text strings in to a list of tokens
news['Clean Text'] = news['Artikel'].apply(process_text)
print(type(news['Artikel']))
```

```python
news.sample(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_transformer = CountVectorizer(analyzer=process_text).fit(news['Clean Text'])

print(len(bow_transformer.vocabulary_)) #Total vocab words
```

```python
#Bag-of-Words (bow) transform the entire DataFrame of text
news_bow = bow_transformer.transform(news['Clean Text'])
```

```python
print('Shape of Sparse Matrix: ', news_bow.shape)
print('Amount of Non-Zero occurences: ', news_bow.nnz)
```

```python
sparsity = (100.0 * news_bow.nnz / (news_bow.shape[0] * news_bow.shape[1]))
print('sparsity: {}'.format(round(sparsity)))
```

```python
#TF-IDF

from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer = TfidfTransformer().fit(news_bow)
news_tfidf = tfidf_transformer.transform(news_bow)
print(news_tfidf)
print(type(news_tfidf))
print(news_tfidf.shape)
```

```python
#Train Naive Bayes Model
from sklearn.naive_bayes import MultinomialNB
import pickle
fakenews_detect_model = MultinomialNB().fit(news_tfidf, news['Fakta/Hoaks'])
```

```python
#save multinominalNB to model

#Model Evaluation
predictions = fakenews_detect_model.predict(news_tfidf)
filename = 'fakenews_model.pkl'
pickle.dump(fakenews_detect_model, open(filename, 'wb'))
print(predictions)
```

```python
from sklearn.metrics import classification_report
print (classification_report(news['Fakta/Hoaks'], predictions))
```

```python
from sklearn.model_selection import train_test_split

news_train, news_test, text_train, text_test = train_test_split(news['Artikel'], news['Fakta/Hoaks'], test_size=0.3)

print(len(news_train), len(news_test), len(news_train) + len(news_test))
print(text_test)
```

```python
#Create Data Pipeline

from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('bow', CountVectorizer(analyzer=process_text)),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', MultinomialNB()),  # train on TF-IDF vectors w/ Naive Bayes classifier
])
pipeline.fit(news_train,text_train)
```

```python
predictions = pipeline.predict(news_test)
```

```python
print(classification_report(predictions,text_test))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(text_test, predictions)
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=['HOAKS', 'FAKTA'], yticklabels=['HOAKS', 'FAKTA'], cmap=plt.cm.Blues, cbar=False)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
