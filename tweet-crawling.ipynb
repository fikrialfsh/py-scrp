```python
#library

import twint
import nest_asyncio
nest_asyncio.apply()
import pandas as pd
```

```python
#konfigurasi

key = input('Enter Keyword: ')
ds = input('Input Date Since') #(yyyy-mm-dd)
du = input('Input Date Until') #(yyyy-mm-dd)
# user = input('Input Username')

c = twint.Config()
c.Search = key
# c.Username = user
c.Limit = 100
c.Since = ds
c.Until = du
c.Lang = "id"
c.Hide_output = True
c.Debug = True
c.Pandas = True
```

```python
#pengambilan data

try:
    print("Mulai Mengunduh Data")
    #konversi output ke dalam format dataframe
    def twint_to_dataframe(columns):
        return twint.output.panda.Tweets_df[columns]
    
    #mengunduh data
    twint.run.Search(c)
    
    #menyimpan data setelah diunduh
    data = twint_to_dataframe(['created_at', 'date', 'user_id', 'username', 'name', 'tweet', 'link'])
    
except Exception as e:
    print(e)
    
print("Tweet telah disimpan")
print("--------")
```
