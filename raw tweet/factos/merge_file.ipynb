```python
import pandas as pd
```

```python
data1 = pd.read_csv('labeled_tweets_covid berasal dari china.csv')
data2 = pd.read_csv('labeled_tweets_vaksin mencegah covid.csv')
data3 = pd.read_csv('labeled_tweets_kasus covid sudah membaik.csv')
data4 = pd.read_csv('labeled_tweets_covid-19 menyebabkan kematian.csv')
data5 = pd.read_csv('labeled_tweets_vaksin tidak wajib dilakukan.csv')
```

```python
data1.drop(data1[data1['Label'] == 0].index, inplace = True)
data1
```

```python
data2.drop(data2[data2['Label'] == 0].index, inplace = True)
data2
```

```python
data3.drop(data3[data3['Label'] == 0].index, inplace = True)
data3
```

```python
data4.drop(data4[data4['Label'] == 0].index, inplace = True)
data4
```

```python
data5.drop(data5[data5['Label'] == 0].index, inplace = True)
data5
```

```python
frames = [data1, data2, data3, data4, data5]
result = pd.concat(frames).drop_duplicates().reset_index(drop=True)
result
```

```python
label = []
for index, row in result.iterrows():
    if row['Label'] == 1:
        label.append('FAKTA')
    else:
        label.append('HOAKS')
result['Label'] = label
result
```

```python
result.to_csv("fact_dataset.csv", index=False)
```
